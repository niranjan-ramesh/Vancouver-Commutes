```python

def dataframer(data):
    all_distances = []
    for i in range(0,len(data)):
        distances = []
        for j in range(0,22):
            if(data[i]['rows'][0]['elements'][j]['status']=='ZERO_RESULTS'):
                distances.append(data[i]['rows'][0]['elements'][j]['status'])
            else:
                distance = int(round(data[i]['rows'][0]['elements'][j]['duration']['value']/60,0))
                distances.append(distance)
        all_distances.append(distances)
    df = pd.DataFrame(all_distances)

    x = pd.read_csv('list.csv')
    x = x['District']

    df['district'] = x

    df['district']
    df = df.replace(to_replace='ZERO_RESULTS', value=0)
    df= df[ ['district'] + [ col for col in df.columns if col != 'district' ]]
    df = df.set_index('district')

    return df
```
