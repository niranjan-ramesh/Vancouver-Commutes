```python
import pandas as pd
import numpy as np

areas = ['Arbutus-Ridge,%20Vancouver|',
 'Downtown,%20Vancouver|',
 'Dunbar-Southlands,%20Vancouver|',
 'Fairview,%20Vancouver|',
 'Grandview-Woodland,%20Vancouver|',
 'Hastings-Sunrise,%20Vancouver|',
 'Kensington-Cedar%20Cottage,%20Vancouver|',
 'Kerrisdale,%20Vancouver|',
 'Killarney,%20Vancouver|',
 'Kitsilano,%20Vancouver|',
 'Marpole,%20Vancouver|',
 'Mount%20Pleasant,%20Vancouver|',
 'Oakridge,%20Vancouver|',
 'Renfrew-Collingwood,%20Vancouver|',
 'Riley%20Park,%20Vancouver|',
 'Shaughnessy,%20Vancouver|',
 'South%20Cambie,%20Vancouver|',
 'Strathcona,%20Vancouver|',
 'Sunset,%20Vancouver|',
 'Victoria-Fraserview,%20Vancouver|',
 'West%20End,%20Vancouver|',
 'West%20Point%20Grey,%20Vancouver|']
```

```python
def url_maker(areas, mode):
    temp_areas = areas
    urls = []
    x = "".join(temp_areas)
    for area in areas:
        url = "https://maps.googleapis.com/maps/api/distancematrix/"\
              "json?origins=" + str(area) +""\
              "&destinations=" + x + ""\
              "&units=metric"\
              "&mode=" + str(mode)+""\
              "&key=<ACCESS_KEY>"
        urls.append(url)
    return urls
```

```python
urls = url_maker(areas,'driving')
```

```python
def url_fetcher():
    import requests
    import json

    full_responses = []
    for url in urls:
        payload={}
        headers = {}
        response = requests.request("GET", url, headers=headers, data=payload)
        wdata = response.text
        data = json.loads(wdata)
        full_responses.append(data)

    return full_responses
```

```python
driving_response = url_fetcher()
```
