```python
def url_maker(areas, mode):
    temp_areas = areas
    urls = []
    x = "".join(temp_areas)
    for area in areas:
        url = "https://maps.googleapis.com/maps/api/distancematrix/"\
              "json?origins=" + str(area) +""\
              "&destinations=" + x + ""\
              "&units=metric"\
              "&mode=" + str(mode)+""\
              "&key=AIzaSyB1h8cmS0_6jkHKRekjoXghIugejRy1QIQ"
        urls.append(url)
    return urls
```

```python
urls = url_maker(areas,'bicycling')
```
